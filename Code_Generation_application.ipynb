```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
from transformers import pipeline
```

```python
!pip install datasets
```

```python
from datasets import load_dataset
dataset = load_dataset("theblackcat102/evol-codealpaca-v1")
```

```python
model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

def generate_code(prompt_text, max_length=100):
  input_ids = tokenizer.encode(prompt_text, return_tensors="pt")

  output = model.generate(input_ids, max_length= max_length, num_return_sequences = 1, no_repeat_ngram_size=2, top_k = 50, top_p = 0.95, temperature = 0.7)
  generate_code = tokenizer.decode(output[0], skip_special_tokens= True)
  return generate_code


prompt = "Generate code to find the sqaure root of a number"
generated_code = generate_code(prompt)
print(generated_code)



```

```python
code_generation_pipeline = pipeline("text-generation", model = model, tokenizer = tokenizer)
generated_code = code_generation_pipeline(prompt, max_length = 100, num_return_sequences=1)
print(generated_code[0]['generated_text'])
```
